## <span style="color: orange;">Loading Dependencies</span>
- `torchvision.transforms` will be used for images transformation
- `pil` will be used for loading and saving images
- `matplotlib.pyplot` will be used for visualizations

```python
from PIL import Image
from torchvision import transforms
import matplotlib.pyplot as plt
```

## <span style="color: orange;">Notebook Utility Methods</span>


```python
def display_image(image, title=None) -> None:
    plt.figure(figsize=(7,7))
    plt.imshow(image)
    plt.axis(False)
    if title:
        plt.title(title)
```

## <span style="color: orange;">Loading an Image Sample</span>


```python
img_src = "assets/sample.png"
image = Image.open(img_src)

display_image(image, title="colored panel from BERSERK Manga")
```

## <span style="color: orange;">Image Transformations Techniques</span>
This section will explore some ***<span style="color: cyan;"> images transformation</span>*** techniques, along with ***<span style="color: cyan;"> Best Practices</span>*** that are applied to image datasets prior to model training. 

Some of the techniques mentioned are:
- `Resize((H, W))`
- `RandomHorizontalFlip()`
- `RandomRotation(val)`
- `GrayScale()`
- `RandomCrop(val)`
- `GaussianBlur(kernel_size, sigma)`
- `Compose([...])`
---

#### Resize the input image to the given size `transforms.Resize((H, W))`
- returns an image with H*W pixels

```python
transform = transforms.Resize((50, 50)) 
resized_image = transform(image)
display_image(resized_image)
```

#### Horizontally flip the given image randomly with `transforms.RandomHorizontalFlip()`

```python
transform = transforms.RandomHorizontalFlip()
rotated_image = transform(image)
display_image(rotated_image)
```

#### Horizontally flip the given image randomly with `transforms.RandomRotation(10)`

```python
transform = transforms.RandomRotation(50)
rnd_rotated_image = transform(image)
display_image(rnd_rotated_image)
```

#### Convert image to grayscale with `transforms.Gratscale()`

```python
transform = transforms.Grayscale()
grayscale_image = transform(image)
display_image(grayscale_image)
```

#### Crop the given image at a random location with `transforms.RandomCrop(500)

```python
transform = transforms.RandomCrop(500)
cropped_image = transform(image)
display_image(cropped_image)
```

#### Blurs image with randomly chosen Gaussian blur with `transforms.GaussianBlur(kernel_size, sigma)`

```python
transform = transforms.GaussianBlur(kernel_size=101, sigma=50) # ressources heavy btw
blurred_image = transform(image)
display_image(blurred_image)
```

#### Convert a PIL Image to tensor and scale the values accordingly `transforms.ToTensor()`
- A Tensor is a mathematical object, represented by an array of components
- This technique is neccessary if you are training a Computer Vision Model
- In most cases, it should be placed last in the image transformation pipeline
- Its crucial that the image is converted to a numerical type, also know as a Tensor
- Machine only understand numbers, vectors and matrixes :) 

```python
transform = transforms.ToTensor()
tensor_image = transform(image)
tensor_image.shape
```

#### Groups Transformation into an organized pipeline with `transforms.Compose([...])`

```python
T = transforms.Compose([
    transforms.Grayscale(),
    transforms.RandomCrop(500),
])

transformed_image = T(image)
display_image(transformed_image)
```
